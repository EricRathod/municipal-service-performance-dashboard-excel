```python
import pandas as pd
import numpy as np
import random

from datetime import datetime, timedelta
```

```python
random.seed(42)
np.random.seed(42)
```

```python
departments = {
    "Public Works": [
        "Pothole Repair",
        "Road Damage",
        "Sidewalk Damage",
        "Snow Removal"
    ],

    "Parks and Recreation": [
        "Tree Maintenance",
        "Park Equipment Damage",
        "Grass Cutting",
        "Trail Maintenance"
    ],

    "Waste Management": [
        "Missed Garbage Pickup",
        "Recycling Issue",
        "Illegal Dumping",
        "Green Bin Issue"
    ],

    "Water Services": [
        "Water Leak",
        "Low Water Pressure",
        "Sewer Backup",
        "Water Quality Concern"
    ],

    "By-law Enforcement": [
        "Noise Complaint",
        "Property Standards",
        "Parking Complaint",
        "Animal Complaint"
    ],

    "Transportation Services": [
        "Streetlight Repair",
        "Traffic Signal Issue",
        "Road Sign Damage",
        "Traffic Concern"
    ]
}

wards = [
    "Ward 1",
    "Ward 2",
    "Ward 3",
    "Ward 4",
    "Ward 5"
]

priorities = [
    "Low",
    "Medium",
    "High",
    "Critical"
]

statuses = [
    "Closed",
    "Open",
    "In Progress"
]

channels = [
    "Web",
    "Phone",
    "Mobile App",
    "Email",
    "In Person"
]
```

```python
records = []

start_date = datetime(2025, 1, 1)

for request_number in range(1, 2001):

    department = random.choice(list(departments.keys()))

    request_type = random.choice(
        departments[department]
    )

    priority = random.choices(
        priorities,
        weights=[25, 45, 24, 6],
        k=1
    )[0]

    status = random.choices(
        statuses,
        weights=[74, 12, 14],
        k=1
    )[0]

    date_opened = start_date + timedelta(
        days=random.randint(0, 364)
    )

    if status == "Closed":

        resolution_ranges = {
            "Low": (3, 15),
            "Medium": (2, 10),
            "High": (1, 7),
            "Critical": (0, 3)
        }

        minimum_days, maximum_days = resolution_ranges[priority]

        resolution_days = random.randint(
            minimum_days,
            maximum_days
        )

        date_closed = date_opened + timedelta(
            days=resolution_days
        )

        citizen_satisfaction = random.choices(
            [1, 2, 3, 4, 5],
            weights=[4, 7, 18, 37, 34],
            k=1
        )[0]

    else:

        resolution_days = np.nan
        date_closed = pd.NaT
        citizen_satisfaction = np.nan

    target_days = {
        "Low": 14,
        "Medium": 10,
        "High": 7,
        "Critical": 3
    }[priority]

    if status == "Closed":

        if resolution_days <= target_days:
            resolved_within_target = "Yes"
        else:
            resolved_within_target = "No"

    else:

        resolved_within_target = "Pending"

    record = {
        "Request ID": f"SR-2025-{request_number:05d}",
        "Department": department,
        "Request Type": request_type,
        "Priority": priority,
        "Status": status,
        "Ward": random.choice(wards),
        "Date Opened": date_opened,
        "Date Closed": date_closed,
        "Channel": random.choice(channels),
        "Citizen Satisfaction": citizen_satisfaction,
        "Resolution Days": resolution_days,
        "Target Days": target_days,
        "Resolved Within Target": resolved_within_target
    }

    records.append(record)
```

```python
municipal_df = pd.DataFrame(records)

municipal_df.head()
```

```python
municipal_df["Opened Month"] = (
    municipal_df["Date Opened"]
    .dt.to_period("M")
    .astype(str)
)

municipal_df["Opened Year"] = (
    municipal_df["Date Opened"]
    .dt.year
)

municipal_df["Opened Quarter"] = (
    municipal_df["Date Opened"]
    .dt.quarter
)

municipal_df["Is Closed"] = np.where(
    municipal_df["Status"] == "Closed",
    "Yes",
    "No"
)

municipal_df.head()
```

```python
print("Rows:", municipal_df.shape[0])
print("Columns:", municipal_df.shape[1])

print("\nMissing Values:")
print(municipal_df.isnull().sum())

print("\nStatus Counts:")
print(municipal_df["Status"].value_counts())

print("\nDepartment Counts:")
print(municipal_df["Department"].value_counts())
```

```python
municipal_df["Date Opened"] = (
    municipal_df["Date Opened"]
    .dt.strftime("%Y-%m-%d")
)

municipal_df["Date Closed"] = (
    pd.to_datetime(municipal_df["Date Closed"])
    .dt.strftime("%Y-%m-%d")
)
```

```python
file_name = "municipal_service_requests_clean.csv"

municipal_df.to_csv(
    file_name,
    index=False
)

print("Dataset successfully created:", file_name)
```
